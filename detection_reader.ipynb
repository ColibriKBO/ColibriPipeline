Code to plot both the lightcurve and show the relevant images of a detection

```python
data_loc = r"D:\LongTermStorage\Pipeline_Testing\ColibriArchive\2024-10-19"

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# get all of the .txt files that start with 'det' in the data_loc directory
files = [f for f in os.listdir(data_loc) if f.endswith('.txt') and f.startswith('det')]
print(len(files), "files found")
# read them all as dataframes with columns ['filename', 'time', 'flux', 'conv_flux]
dfs = []
for f in files:
    try:
        df = pd.read_csv(os.path.join(data_loc, f), delim_whitespace=True, skiprows=lambda x: x< 19, names=['filename', 'time', 'flux', 'conv_flux'])
        dfs.append(df)
    except:
        pass

print(len(dfs), "dataframes found")
```

```python
detection_number = 18

specific_df = dfs[detection_number]
times = specific_df['time'].values
flux = specific_df['flux'].values

# plot the flux vs time
plt.plot(times, flux)
plt.xlabel('Time (s)')
plt.ylabel('Flux')
plt.title('Flux vs Time for ' + specific_df['filename'].values[0])
plt.grid()
plt.show()
```
